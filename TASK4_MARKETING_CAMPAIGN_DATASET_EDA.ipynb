```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("marketing_campaign.csv", sep="\t")
```

```python
df.info()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Income"].fillna(df["Income"].median(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)

print(df.shape)
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format="%d-%m-%Y")
```

```python
df["Age"] = 2026 - df["Year_Birth"]
```

```python
df["Total_Spending"] = (
    df["MntWines"]
    + df["MntFruits"]
    + df["MntMeatProducts"]
    + df["MntFishProducts"]
    + df["MntSweetProducts"]
    + df["MntGoldProds"]
)
```

```python
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=20, color="skyblue")

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Income"], bins=20, color="orange")

plt.title("Income Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Total_Spending"], color="green")

plt.title("Customer Spending")

plt.show()
```

```python
plt.figure(figsize=(15,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap="coolwarm",
    annot=False
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
channels = [
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases"
]

df[channels].sum().plot(
    kind="bar",
    figsize=(7,5),
    color=["blue","orange","green"]
)

plt.title("Purchases by Channel")

plt.ylabel("Number of Purchases")

plt.show()
```

```python
df["Response"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.ylabel("")

plt.title("Campaign Response")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Income",
    y="Total_Spending",
    data=df
)

plt.title("Income vs Total Spending")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Education",
    order=df["Education"].value_counts().index
)

plt.xticks(rotation=45)

plt.title("Education Distribution")

plt.show()
```

```python
plt.figure(figsize=(9,5))

sns.countplot(
    data=df,
    x="Marital_Status",
    order=df["Marital_Status"].value_counts().index
)

plt.xticks(rotation=45)

plt.title("Marital Status")

plt.show()
```

```python
products = [
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "MntGoldProds"
]

df[products].sum().plot(
    kind="bar",
    figsize=(9,5)
)

plt.title("Product-wise Spending")

plt.ylabel("Amount")

plt.show()
```

```python
plt.figure(figsize=(9,5))

sns.barplot(
    data=df,
    x="Education",
    y="Total_Spending",
    estimator=np.mean
)

plt.xticks(rotation=45)

plt.title("Average Spending by Education")

plt.show()
```
